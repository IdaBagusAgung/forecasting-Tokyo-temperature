```python
import pandas as pd
import matplotlib.pyplot as plt
```

# WINTER

```python
ann_winter = pd.read_csv('HASIL_PREDICT/15_ANN_WINTER.csv')
rrn_winter = pd.read_csv('HASIL_PREDICT/15_RNN_WINTER.csv')
lstm_winter = pd.read_csv('HASIL_PREDICT/15_LSTM_WINTER.csv')
```

```python
plt.figure(figsize=(15, 6))

# Subplot untuk prediksi
plt.subplot(1, 2, 1)
plt.plot(ann_winter, label='ANN', marker='o', color='blue', linewidth=2)
plt.plot(rrn_winter, label='RNN', marker='x', color='orange', linewidth=1.5)
plt.plot(lstm_winter, label='LSTM', marker='^', color='green', linewidth=1)
plt.xlabel('Hari ke-')
plt.ylabel('Nilai Prediksi')
plt.title('Perbandingan Prediksi Cuaca')
plt.legend()
plt.grid(True)
plt.ylim(6, 16)

plt.tight_layout()
plt.show()
```

# RMSE PERBANDINGAN

## WINTER

```python
import pandas as pd
```

```python
rmse_ann_winter = pd.read_csv('HASIL/ANN_WINTER.csv')
rmse_rnn_winter = pd.read_csv('HASIL/RNN_WINTER.csv')
rmse_lstm_winter = pd.read_csv('HASIL/LSTM_WINTER.csv')
```

```python
rmse_ann_winter = rmse_ann_winter['RMSE']
rmse_rnn_winter = rmse_rnn_winter['RMSE']
rmse_lstm_winter = rmse_lstm_winter['RMSE']
```

```python
# Label untuk sumbu x
model_labels = ['Model 1', 'Model 2', 'Model 3']

# Membuat plot
plt.figure(figsize=(12, 6))
plt.plot(rmse_ann_winter, label='ANN')
plt.plot(rmse_rnn_winter, label='RNN')
plt.plot(rmse_lstm_winter, label='LSTM')

# Memberikan label pada sumbu x dan y
plt.xticks(range(len(model_labels)), model_labels)  # Mengatur label sumbu x
plt.ylabel('Nilai RMSE')

# Memberikan judul plot
plt.title('Perbandingan RMSE WINTER')

# Menampilkan legend
plt.legend()

# Menampilkan plot
plt.show()
```
