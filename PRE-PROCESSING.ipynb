# CUT DATA SUMMER/WINTER/AUTUMN/SPRING

```python
data_clean = pd.read_csv('DATASET/DATA_TOKYO_BERSIH.csv')
```

```python
data_clean.info()
```

```python
season = data_clean['season'].unique()
print(season)
```

```python
# Mengelompokkan data berdasarkan musim
grouped = data_clean.groupby('season')

# Menyimpan setiap DataFrame musim ke file CSV yang berbeda
for name, group in grouped:
    group.to_csv(f'DATASET/SEASON/data_{name}.csv', index=False)
```
